# Merging Data 

```python
import pandas as pd
import matplotlib.pyplot as plt

chip = pd.read_csv("fast food.csv", sep=",")


zipList = []
for i in range(837):#test for the zip code 15203 since it comes up 3 times in 0-15
    x = chip.iloc[i, 5]
    zipList.append(x)

d = dict()
for c in zipList:
    if c not in d:
        d[c] = 1
    else:
        d[c] = d[c] + 1


column_names=["Zip Code","Fast Food Resterants"]

#df = pd.DataFrame(d, index=['i',], columns=column_names)
df_FastFood = pd.DataFrame(list(d.items()), columns=column_names)
df_FastFood = df_FastFood.sort_values('Fast Food Resterants', ascending=False)
```

```python
import re

def getzc(incidentLocation):#this method is used to take the information under the Incident Location column and get the zip code
    zc = r'\b\d{5}\b' #matches 5 digits at the end of a string
    y = re.search(zc, incidentLocation)
    
    if y:#zip code is found
        return y.group(0)
    else:#no zip code is found
        return None
    
chip = pd.read_csv("Police Incident Blotter data.csv", sep=",")



zipList = []
for i in range(312145):
    x = chip.iloc[i, 4]
    xzc = getzc(x)
    #print(xzc) #will print either a 5 digit zip code or "None"
    zipList.append(xzc)
    
d = dict()
for c in zipList:
    if c not in d:
        d[c] = 1
    else:
        d[c] = d[c] + 1
        
print("this is the number of crimes per zip code from January 1st, 2016, to Febuary 18th, 2023")
print(d)
print("there are ", len(d), " different zip codes in this data set")

        

column_names=["Zip Code","Crime Rates"]

#df = pd.DataFrame(d, index=['i',], columns=column_names)
df_CrimeRates = pd.DataFrame(list(d.items()), columns=column_names)
df_CrimeRates= df_CrimeRates.sort_values('Crime Rates', ascending=False)
                    
```

```python
chip3 = pd.read_csv("School Districts.csv", sep="\,")
zipList = []
for i in range(101):#test for the zip code 15203 since it comes up 3 times in 0-15
    x = chip3.iloc[i, 5]
    zipList.append(x)
    
dic2 = dict()
for c in zipList:
    if c not in dic2:
        dic2[c] = 1
    else:
        dic2[c] = dic2[c] + 1
        
column_names=["Zip Code","K-12 Schools"]

df_Schools= pd.DataFrame(list(dic2.items()), columns=column_names)

df_Schools = df_Schools .sort_values("K-12 Schools", ascending=False)
```

```python
# Data cleaning to prepare for joining data

# dropping NaA values
df_CrimeRates.dropna(subset=['Zip Code'], inplace=True)
df_Schools.dropna(subset=['K-12 Schools'], inplace=True)
df_FastFood.dropna(subset=['Fast Food Resterants'], inplace=True)

df_CrimeRates['Zip Code'] = df_CrimeRates['Zip Code'].astype(int)
joint_df = df_CrimeRates.merge(df_Schools, on='Zip Code', how='right')

joint_df=joint_df.merge(df_FastFood, on='Zip Code', how='left')



joint_df.dropna(subset=['Fast Food Resterants'], inplace=True)
joint_df.dropna(subset=['K-12 Schools'], inplace=True)


joint_df.sort_values("K-12 Schools", ascending=False, inplace=True)
joint_df.head(30)

# Weight scales: Food: 40, Crime:35, School: 25

# Scoring Scales
# schools: max(15) min(1)
# resterants: max(46) min(1)
# Crime Rate: max(34146) min(3970)


joint_df.sort_values("Crime Rates", ascending=False, inplace=True)
joint_df.head(30)

joint_df['Crime Rate Score (weighed 0.4)'] = (1 -joint_df['Crime Rates'] /  34146 )*100*0.4
joint_df["K-12 Schools (weighed 0.35)"] = (joint_df["K-12 Schools"] / 15 )*100 *0.35
joint_df["Fast Food Resterants (weighed 0.25)"] = (joint_df["Fast Food Resterants"] / 46)*100 *0.25

#Joining the dataframes together 

joint_df["Total Score"] = (joint_df['Crime Rate Score (weighed 0.4)'])+(joint_df["K-12 Schools (weighed 0.35)"])+(joint_df["Fast Food Resterants (weighed 0.25)"])

joint_df.sort_values("Total Score", ascending=False, inplace=True)

joint_df.head(30)
```
